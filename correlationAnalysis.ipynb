```python
import pandas as pd
```

```python
df = pd.read_csv("#dataset_path")
dataset = df.copy()
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
labels = {  'none' : 0,
            'weakest_link' : 0,
           'elite_player' : 1,
           'urgent_followup_is_needed': 1, 
           'followup_is_needed': 1}
# create grouped label column
dataset['Label_Category_Num'] = dataset['label'].map(lambda x: labels[x])
dataset.head()
```

```python
df_attribute = dataset.iloc[:,7:48]
df_label = dataset.iloc[:,[-1]]
df_for_corr = pd.concat([df_attribute, df_label], axis=1)
df_for_corr.head()
#correlations = df_for_corr.corr()
corr = df_attribute.corrwith(df_label['Label_Category_Num'])
print(corr.shape)
print(type(corr))
```

```python
df_attribute = dataset.iloc[:,7:48]
df_label = dataset.iloc[:,[-1]]
df_for_corr = pd.concat([df_attribute, df_label], axis=1)
df_for_corr.head()
#correlations = df_for_corr.corr()
corr_spearman = df_attribute.corrwith(df_label['Label_Category_Num'],method="spearman")
s1_spearman = corr_spearman.sort_values(ascending=True)
print(s1_spearman)
s1_spearman.to_excel("pos1_spearman.xlsx") 
```
